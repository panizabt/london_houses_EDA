```python
import warnings
warnings.filterwarnings("ignore")
sns.set_style()
```

```python
data = pd.read_csv("london_houses_dataset.csv")
df = pd.DataFrame(data)
```

```python
def count_plot(col, label, rotation=0, size=(12, 6)):
    plt.figure(figsize=size)
    plt.title(f'Count by house {label}', fontsize=20)
    sns.countplot(x=data[col], palette='viridis')
    plt.xlabel(label, fontsize=16)
    plt.xticks(rotation=rotation)
    plt.ylabel('Count', fontsize=16)
```

```python
count_plot("Building Age", "Price", rotation=100, size=(20, 8))
```

```python
count_plot("Building Age", "Floors", rotation=90, size=(20, 8))
```

```python
count_plot("Floors", "Square Meters", rotation=90, size=(16, 5))
```

```python
count_plot("Address", "Neighborhood", rotation=90, size=(200, 50))
```

```python
count_plot("Building Status", "Building Age", rotation=90, size=(16, 8))
```

```python
count_plot("Property Type", "Heating Type", rotation=90, size=(16, 5))
```

```python
def box_plot(col, label):
    plt.figure(figsize=(10, 6))
    sns.boxplot(data[col], palette="viridis")
    plt.title(label, fontsize=20)
    plt.ylabel(label, fontsize=16)
```

```python
box_plot("Building Age", "Price")

```

```python
box_plot("Building Age", "Floors")
```

```python
box_plot("Floors", "Square Meters")
```

```python
box_plot("Floors", "floors")
```

```python
def histo_plot(col, label):
    plt.figure(figsize=(10, 6))
    sns.histplot(data[col], palette="viridis")
    plt.title(label, fontsize=30)
    plt.xlabel(label, fontsize=30)
```

```python
histo_plot("Floors", "floors")
```

```python
histo_plot("Building Age", "Building Age")
```

```python
histo_plot("Property Type", "Square Meters")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df["Price (£)"], bins=30, kde=True, color="blue")
plt.title("Distribution of House Prices", fontsize=16)
plt.xlabel("Price (£)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.show()
```

```python
neighborhood_prices = df.groupby("Neighborhood")["Price (£)"].mean().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
neighborhood_prices.plot(kind="bar", color="green")
plt.title("Average House Price by Neighborhood", fontsize=16)
plt.xlabel("Neighborhood", fontsize=12)
plt.ylabel("Average Price (£)", fontsize=12)
plt.xticks(rotation=0)
plt.show()
```

```python
bedroom_prices = df.groupby("Bedrooms")["Price (£)"].mean().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
bedroom_prices.plot(kind="bar", color="pink")
plt.title("Average House Price by Bedrooms", fontsize=16)
plt.xlabel("Bedrooms", fontsize=12)
plt.ylabel("Average Price (£)", fontsize=12)
plt.xticks(rotation=0)
plt.show()
```

```python
correlation_matrix = df[["Bedrooms", "Bathrooms", "Square Meters", "Building Age", "Price (£)"]].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix", fontsize=16)
plt.show()
```

```python
property_prices = df.groupby("Property Type")["Price (£)"].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
property_prices.plot(kind="bar", color="coral")
plt.title("Average Price by Property Type", fontsize=16)
plt.xlabel("Property Type", fontsize=12)
plt.ylabel("Average Price (£)", fontsize=12)
plt.xticks(rotation=0)
plt.show()
```

```python
garden_price = df.groupby("Garden")["Price (£)"].mean()
plt.figure(figsize=(8, 5))
garden_price.plot(kind="bar", color=["lightgreen", "pink"])
plt.title("Effect of Garden Presence on House Price", fontsize=16)
plt.xlabel("Garden", fontsize=12)
plt.ylabel("Average Price (£)", fontsize=12)
plt.xticks(rotation=0)
plt.show()
```

```python
garage_price = df.groupby("Garage")["Price (£)"].mean()
plt.figure(figsize=(8, 5))
garage_price.plot(kind="bar", color=["lightgreen", "lightblue"])
plt.title("Effect of Garage on House Price", fontsize=16)
plt.xlabel("Garage", fontsize=12)
plt.ylabel("price (£)", fontsize=12)
plt.xticks(rotation=0)
plt.show()
```

```python
heating_prices = df.groupby("Heating Type")["Price (£)"].mean()
plt.figure(figsize=(10, 6))
heating_prices.plot(kind="bar", color="purple")
plt.title("Price by Heating Type", fontsize=16)
plt.xlabel("Heating Type", fontsize=12)
plt.ylabel("Price (£)", fontsize=12)
plt.xticks(rotation=0)
plt.show()
```

```python
balcony_prices = df.groupby("Balcony")["Price (£)"].mean()
plt.figure(figsize=(10, 6))
balcony_prices.plot(kind="bar", color="purple")
plt.title("Price by Balcony Type", fontsize=16)
plt.xlabel("Balcony Type", fontsize=12)
plt.ylabel("price (£)", fontsize=12)
plt.xticks(rotation=0)
plt.show()
```

```python
view_price = df.groupby("View")["Price (£)"].mean()
plt.figure(figsize=(10, 6))
view_price.plot(kind="bar", color="pink")
plt.title("Price by View Type", fontsize=16)
plt.xlabel("View Type", fontsize=12)
plt.ylabel("Price(£)", fontsize=12)
plt.xticks(rotation=0)
plt.show()
```

```python
Materials_prices = df.groupby("Materials")["Price (£)"].mean()
plt.figure(figsize=(10, 6))
Materials_prices.plot(kind="bar", color="blue")
plt.title("Average Price by View Type", fontsize=16)
plt.xlabel("Materials Type", fontsize=12)
plt.ylabel("Average Price (£)", fontsize=12)
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x="Building Age", y="Price (£)", data=df, hue="Neighborhood", palette="viridis", alpha=0.7)
plt.title("Building Age vs House Price", fontsize=16)
plt.xlabel("Building Age", fontsize=12)
plt.ylabel("Price (£)", fontsize=12)
plt.show()
```
